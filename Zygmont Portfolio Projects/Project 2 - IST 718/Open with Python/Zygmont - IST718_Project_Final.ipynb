Pre-processing

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer

# Convert 'Cancelled' column to boolean type
flights21['Cancelled'] = flights21['Cancelled'].astype(bool)

# Drop unnecessary columns
drop_columns = [
    'FlightDate', 'Flight_Number_Marketing_Airline', 'Flight_Number_Operating_Airline', 'Tail_Number',
    'CRSDepTime', 'DepTime', 'ArrTime', 'CRSArrTime', 'Operated_or_Branded_Code_Share_Partners',
    'DivAirportLandings', 'Marketing_Airline_Network', 'DOT_ID_Marketing_Airline', 'IATA_Code_Marketing_Airline',
    'Operating_Airline', 'DOT_ID_Operating_Airline', 'IATA_Code_Operating_Airline', 'OriginAirportID',
    'OriginAirportSeqID', 'OriginCityMarketID', 'OriginCityName', 'OriginState', 'OriginStateFips',
    'OriginStateName', 'OriginWac', 'DestAirportID', 'DestAirportSeqID', 'DestCityMarketID', 'DestCityName',
    'DestState', 'DestStateFips', 'DestStateName', 'DestWac', 'DepartureDelayGroups', 'ArrDelay', 'ArrDel15',
    'ArrivalDelayGroups', 'ArrTimeBlk', 'DistanceGroup', 'Region', 'DepDel15', 'DepartureDelayGroups',
    'TaxiOut', 'WheelsOff', 'WheelsOn', 'TaxiIn', 'ArrivalDelayGroups', 'ArrDelayMinutes'
]
flights21_cleaned = flights21.drop(columns=drop_columns)

# Impute missing numerical values
numerical_columns = ['DepDelayMinutes', 'AirTime', 'Distance', 'CRSElapsedTime', 'ActualElapsedTime', 'DepDelay']
imputer_numerical = SimpleImputer(strategy='mean')  # Using mean for numerical values
flights21_cleaned[numerical_columns] = imputer_numerical.fit_transform(flights21_cleaned[numerical_columns])

# Impute missing categorical values with the mode
categorical_columns = ['Airline', 'Origin', 'Dest', 'DayOfWeek', 'Month', 'DayofMonth']
imputer_categorical = SimpleImputer(strategy='most_frequent')  # Using most frequent value for categorical columns
flights21_cleaned[categorical_columns] = imputer_categorical.fit_transform(flights21_cleaned[categorical_columns])

print(flights21_cleaned.isnull().sum())

# Encode categorical variables using LabelEncoder
label_encoder = LabelEncoder()
flights21_cleaned['Airline'] = label_encoder.fit_transform(flights21_cleaned['Airline'])
flights21_cleaned['Origin'] = label_encoder.fit_transform(flights21_cleaned['Origin'])
flights21_cleaned['Dest'] = label_encoder.fit_transform(flights21_cleaned['Dest'])
flights21_cleaned['DayOfWeek'] = label_encoder.fit_transform(flights21_cleaned['DayOfWeek'])
flights21_cleaned['Month'] = label_encoder.fit_transform(flights21_cleaned['Month'])
flights21_cleaned['DayofMonth'] = label_encoder.fit_transform(flights21_cleaned['DayofMonth'])

# Select features (X) and target variable (y)
features = ['Airline', 'Origin', 'Dest', 'Diverted', 'DepDelayMinutes', 'AirTime', 'Distance', 'Year', 'Quarter',
            'Month', 'DayofMonth', 'DayOfWeek']
X = flights21_cleaned[features]
y = flights21_cleaned['Cancelled']

# Standardize numerical features
scaler = StandardScaler()
X[['DepDelayMinutes', 'AirTime', 'Distance']] = scaler.fit_transform(X[['DepDelayMinutes', 'AirTime', 'Distance']])

# Preprocessed data
X_cleaned = X
y_cleaned = y
```

```python
# Check for any missing values in the dataframe
print(flights21_cleaned.isnull().sum())

# Check specifically for rows with missing values
missing_rows = flights21_cleaned[flights21_cleaned.isnull().any(axis=1)]
print(missing_rows)
```

Modeling

```python
from sklearn.model_selection import train_test_split

# train-test split to maintain class distribution
X_train, X_test, y_train, y_test = train_test_split(X_cleaned, y_cleaned, test_size=0.3, random_state=42, stratify=y_cleaned)

print("Training set class distribution:")
print(y_train.value_counts())
```

```python
from sklearn.utils import resample

# Separate majority and minority class
df_majority = X_train[y_train == False]
df_minority = X_train[y_train == True]

# Check if df_minority is empty before resampling
if len(df_minority) > 0:
    # Upsample the minority class if it's not empty
    df_minority_upsampled = resample(df_minority,
                                     replace=True,
                                     n_samples=len(df_majority),
                                     random_state=42)


    X_train_upsampled = pd.concat([df_majority, df_minority_upsampled])
    y_train_upsampled = pd.concat([y_train[y_train == False], y_train[y_train == True]])
else:

    print("Minority class is empty. Cannot perform upsampling.")
    X_train_upsampled = df_majority  # Use majority class as is
    y_train_upsampled = y_train[y_train == False]


print("Class distribution after oversampling:")
print(y_train_upsampled.value_counts())
```

```python
from sklearn.linear_model import LogisticRegression

# Logistic Regression with class weights
lr_classifier = LogisticRegression(class_weight='balanced', random_state=42)
lr_classifier.fit(X_train, y_train)

# Make predictions
y_pred_lr = lr_classifier.predict(X_test)

# Evaluate the model
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
accuracy_lr = accuracy_score(y_test, y_pred_lr)
print("Logistic Regression - Accuracy:", accuracy_lr)
print(classification_report(y_test, y_pred_lr))
print(confusion_matrix(y_test, y_pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

# Decision Tree Classifier with Class Weights
dt_classifier = DecisionTreeClassifier(class_weight='balanced', random_state=42)
dt_classifier.fit(X_train, y_train)

# Make predictions
y_pred_dt = dt_classifier.predict(X_test)

# Evaluate the Decision Tree model
accuracy_dt = accuracy_score(y_test, y_pred_dt)
print("Decision Tree - Accuracy:", accuracy_score(y_test, y_pred_dt))
print(classification_report(y_test, y_pred_dt))
```
